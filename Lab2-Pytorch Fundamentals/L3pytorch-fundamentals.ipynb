Tensor datatypes

PyTorch provides a rich set of tensor data types (dtype) that define how numerical values are stored and processed. These data types play a crucial role in determining memory usage, computational speed, and numerical precision in deep learning models.

CPU vs GPU Tensor Types:

PyTorch tensors can run on both CPU and GPU.
* GPU-based tensors are optimized for parallel computation and often use:
torch.device("cuda") for GPU acceleration
* CPU tensors use:
torch.device("cpu") (default)

Default Data Type
* torch.float32 (or torch.float)


Floating-Point Data Types
* torch.float16 (torch.half)
* torch.float32 (torch.float)
* torch.float64 (torch.double)

Why Different Data Types Exist?

Different data types exist to optimize the trade-off between:

Precision (Accuracy)
* Higher precision = more accurate representation
* Reduces numerical errors in computations

Performance (Speed & Memory)
* Lower precision = faster computation
* Uses less memory, enabling larger models and batches

```python
# Default datatype for tensors is float32
float_32_tensor = torch.tensor([2.0, 4.0, 6.0],
                               dtype=None, # defaults to None, which is torch.float32 or whatever datatype is passed
                               device=None, # defaults to None, which uses the default tensor type
                               requires_grad=False) # if True, operations performed on the tensor are recorded 

float_32_tensor.shape, float_32_tensor.dtype, float_32_tensor.device
```

```python
float_16_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=torch.float16) # torch.half would also work

float_16_tensor.dtype
```

Tensor Operations

In deep learning, all types of data such as images, text, audio, videos, and numerical information are first converted into tensors (multi-dimensional arrays of numbers).

A neural network learns by performing large-scale mathematical operations on these tensors, allowing it to detect patterns, relationships, and meaningful structures in the data. During training, the model may execute millions of computations to optimize its performance.

Common Tensor Operations

Some of the most frequently used tensor operations include:

* Addition
* Subtraction
* Element-wise multiplication
* Division
* Matrix multiplication

```python
# Create a tensor of values and add a number to it
tensor = torch.tensor([1, 2, 3])
tensor + 10
```

```python
# Multiply it by 10
tensor * 10
```

```python
tensor
```

```python
# Subtract and reassign
tensor = tensor - 10
tensor
```

```python
# Can also use torch functions
torch.multiply(tensor, 10)
```

```python
tensor
```

```python
tensor = tensor + 10
# Element-wise multiplication (each element multiplies its equivalent, index 0->0, 1->1, 2->2)
print(tensor, "*", tensor)
print("Equals:", tensor * tensor)
print("Equals:", torch.multiply(tensor, tensor))
```

```python
# Let's create a tensor and perform element-wise multiplication and matrix multiplication on it.
tensor = torch.tensor([1, 2, 3])

# Element-wise matrix multiplication
print(tensor * tensor)

# Matrix multiplication
print(torch.matmul(tensor, tensor))
```

One of the most common errors in deep learning (shape errors)

```python
# Shapes need to be in the right way  
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B.T) # (this will error)
```

```python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output) 
print(f"\nOutput shape: {output.shape}")
```

Reshaping

* torch.reshape(input, shape) is used to change the shape of a tensor without altering its data. The total number of elements must remain the same. It can also be called using torch.Tensor.reshape(), and it may return either a view or a copy depending on memory layout.

```python
# Create a tensor
import torch
x = torch.arange(1., 9.)
print(x, x.shape)

# Add an extra dimension
x_reshaped = x.reshape(1, 8) # try other shapes, it will produce an error.
x_reshaped, x_reshaped.shape
```

```python
x, x.shape
```

View

* Tensor.view(shape) is used to change the shape of a tensor without copying data. It creates a new view of the same tensor, meaning it shares the same memory as the original tensor.

```python
# Change view (keeps same data as original but changes view)
z = x.view(1, 8)
z, z.shape
```

```python
z = x.view(2,4)
z, z.shape
```

```python
x, x.shape
```

```python
x = torch.arange(6)

a = x.view(2, 3)
b = x.reshape(2, 3)

a[0, 0] = 100
print(x)
```

Stack

* torch.stack(tensors, dim=0) is used to combine multiple tensors into one tensor by adding a new dimension. All tensors must have the same shape.

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) 
x_stacked
```

```python
# Stacking using dim=1
x_stacked = torch.stack([x, x, x, x], dim=1) 
x_stacked
```

Squeeze and Unsqueeze funtions.

* squeeze() → removes extra 1 dimension
* unsqueeze() → adds a new 1 dimension

```python
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

permute

* torch.permute() is used to reorder the dimensions (axes) of a tensor without changing its data values. It returns a new view of the tensor with a different dimension arrangement, which is commonly used for reshaping data in deep learning (e.g., changing image format from (C, H, W) to (H, W, C)).

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.shape)

y = x.permute(1, 0)

print(y)
print(y.shape)
```

Indexing

```python
# Create a tensor 
import torch
x = torch.arange(1, 10).reshape(1, 3, 3)
x, x.shape

# Let's index bracket by bracket
print(f"First square bracket:\n{x[0]}") 
print(f"Second square bracket: {x[0][0]}") 
print(f"Third square bracket: {x[0][0][0]}")
```

```python
print(x[:, 0]) # Get all values of 0th dimension and the 0 index of 1st dimension
print(x[:, :, 1])# Get all values of 0th & 1st dimensions but only index 1 of 2nd dimension
print(x[:, 1, 1]) # Get all values of the 0 dimension but only the 1 index value of the 1st and 2nd dimension
print(x[0, 0, :]) # same as x[0][0]
```

Pytorch Tensor to NumPy arrays and vice versa

* torch.from_numpy(ndarray) - NumPy array -> PyTorch tensor.
* torch.Tensor.numpy() - PyTorch tensor -> NumPy array.

```python
# NumPy array to tensor
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7) # create a tensor of ones with dtype=float32
numpy_tensor = tensor.numpy() # will be dtype=float32 unless changed
tensor, numpy_tensor
```
